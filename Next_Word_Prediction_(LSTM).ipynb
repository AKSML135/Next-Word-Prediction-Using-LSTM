## First we will be Importing tensorflow and Tokenizer - The main reason using Tokenizer is to map each word with a number
## Example - I want apple
- I - 1, want -2 , apple - 3

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
token = Tokenizer()
token.fit_on_texts([text])
```

```python
# token.word_index will give us the mapped number to a word
token.word_index
```

## MAKING OUR DATA FROM INPUT TEXT
Now for each sentences we will be first split  on the basis of new line .
After that we are using texts_to_sequences method to create a list of numbers that are mapped to each number
Example - I live in Mumbai - [12,23,219,1]

finally with the help of for loop we are making our dataset which we have discussed in the readme file -
[12,23] , [12,23,219] , [12,23,219,1] for sentence 1 so that we can further split it into X and Y

```python
input_sequences = []
for sentence in text.split('\n'):
  tokenized_sentence = token.texts_to_sequences([sentence])[0]

  for i in range(1, len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
  print(input_sequences)
```

## 0 PADDING TO INPUT EMBEDDING
Now we need padding as these are our embedding and all the embedding should be of same dimension. To acheive that we will be using pad_sequences function of keras which will perform 0 padding to that start (pre) of each of our list of embeddings.
To make embedding of same dim. we are finding the maximum length sentence of our text and use this as our length (dimension) of each embedding list.And this makes sense as well !

```python
max_len = max([len(x) for x in input_sequences])
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding = 'pre')
```

```python
padded_input_sequences
```

## MAKING INPUT FEATURES AND TARGET
We will spilt our list now to X and Y, simple logic take all columns except last as X and last column as Y

```python
X = padded_input_sequences[:,:-1]
Y = padded_input_sequences[:,-1]
```

## ONE HOT ENCODING - TARGET VARIABLE
Now we will convert our Y into one hot encoding using to_categorical function of tensorflow. The total number of classes that are present in Y is equal to total number of words in our text. which is 417+1 as indexing starts from 1.

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(Y, num_classes = 418)
```

```python
y
```

## DEFINING MODEL PARAMETERS

Now The main part, defining our neural network. Our NN will contain 3 layers.
1 is Embedding layer that will convert each word of our list to a vector of 100 dimension. Total words in our vocab is 418 and total number of rows are 302.
2 is LSTM layer which will output in 200 dimension
and then a dense layer which will again contain the probability of each word so the dimension should be equal to the total word in vocab and the softmax activation will give us probab for each word

While Compliling we are using categorical_crossentropy as our loss function which is used for categorical data , optimiser as adam and metrics as accuracy

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, GRU
```

```python
model = Sequential()
model.add(Embedding(418, 100, input_length = 302))
model.add(LSTM(200))
model.add(Dense(418, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model2 = Sequential()
model2.add(Embedding(418, 100, input_length = 302))
model2.add(GRU(200))
model2.add(Dense(418, activation = 'softmax'))
```

```python
model2.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model2.summary()
```

## TRAINING
Now we are fitting our X and Y and doing forward and backward propagation 100 times so that the weights are properly optimised.

P.S -  Notice that after epoch 40 are getting same accuracy . we could use early stoppage here

```python
model.fit(X,y, epochs = 100)
```

## PREDICTION
Finally we are testing our model by predicting next word 10 times .
Ofcourse first we need to map our text to a number and then 0 pad it and then pass it to the model

```python
import numpy as np

pred_text = 'Hindu'

for i in range(10):
  token_pred_text = token.texts_to_sequences([pred_text])[0]
  padded_token_pred_text = pad_sequences([token_pred_text], maxlen = 302, padding = 'pre')
  pos = np.argmax(model.predict(padded_token_pred_text))

  for word,index in token.word_index.items():
    if index == pos:
      pred_text= pred_text+ ' '+ word
      print(pred_text)
```
